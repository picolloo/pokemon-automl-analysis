```python
import pandas as pd
```

```python
df = pd.read_csv('data/pokemon.csv')
df.head()
```

```python
!pip install pycaret
```

```python
from pycaret import classification
```

```python
to_remove = ['abilities','japanese_name', 'name', 'pokedex_number', 'generation']
```

```python
ml_setup = classification.setup(data=df, 
                                  target='is_legendary', 
                                  ignore_features=to_remove,
                                  train_size=0.8, 
                                  session_id=1234
                                )
```

```python
best_model = classification.compare_models(fold=5)
```

```python
classification.predict_model(best_model)
```

```python
classification.evaluate_model(best_model)
```
